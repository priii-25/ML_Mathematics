```python
a=np.array([2,3])
b=np.array([4,4])
res=np.dot(a,b)
print(res)
```

```python
#proj
a=np.array([2,5])
v=np.array([8,-6]) #shape of vecs to be same shape
#magni = np.linalg.norm(v)
magni=np.sqrt(np.sum(v**2))
re=(np.dot(a,v/magni**2))*v
print(re)
```
